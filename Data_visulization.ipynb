# plotting in python


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
kashti=sns.load_dataset("titanic")
kashti
```

```python
kashti.info()   # first check the info of data set
```

```python
kashti.isnull()  # to seee the null value in the whole dataset
```

```python
kashti.isnull().sum()     # to seee who many null has a each data set
```

```python
kashti.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x="survived",data=kashti)
```

```python
sns.countplot(x="survived", hue="sex",data=kashti)
```

```python
plt.scatter(data=kashti,x="age",y="fare")
```

```python
# seaborn library
sns.scatterplot(data=kashti,x="age",y="fare",hue="pclass")
```

```python
sns.scatterplot(data=kashti,x="sex",y="fare",hue="pclass")
```

```python
sns.barplot(data=kashti,x="sex",y="fare",hue="pclass")
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist()
```

```python
kashti.hist(bins=10,figsize=(9,7))
```

```python
kashti.hist(bins=10,figsize=(9,7),grid="false")
```

```python
sns.boxplot(data=kashti,x="sex",y="fare",hue="parch")
```

```python
sns.boxplot(data=kashti,x="sex",y="fare",hue="who")    # after the hue we give the any columns name to see the different plot
```

```python
sns.boxenplot(data=kashti,x="sex",y="fare",hue="who") 
```

```python
sns.violinplot(data=kashti,x="sex",y="fare",hue="who") 
```

```python
g=sns.FacetGrid(kashti,hue="survived",col="pclass",margin_titles=True, 
                palette={1:"seagreen",0:"gray"})
g=g.map(plt.scatter,"fare","age",edgecolor="w").add_legend();
```

```python
sns.stripplot(data=kashti,x="sex",y="fare",hue="who") 
```
